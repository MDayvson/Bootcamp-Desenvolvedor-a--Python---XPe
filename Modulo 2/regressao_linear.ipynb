```python
import matplotlib.pyplot as plt
```

```python
import numpy as np

x = [-1, -0.77777778, -0.5555556, -0.3333333, -0.11111111,
     0.111111, 0.33333333, 0.5555556, 0.77777778, 1]

y = [-1.139556201, -0.57177999, -0.21697092, 0.542599, 0.49406657,
     1.1472239, 1.642228553, 2.17499824, 2.64773614, 2.95684202]

```

```python
# plot dos dados

plt.figure(figsize=(10,5))
plt.plot(x, y, 'o', label='dados originais')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

```python
# estimar uma função do tipo y= a*x + b
# achar os valores de a e b

x, y = np.array(x).reshape(-1, 1), np.array(y).reshape(-1, 1)

# adicionando bias: para estimar o termo b
x = np.hstack((x, np.ones(x.shape)))

# estimando a e b
beta = np.linalg.pinv(x).dot(y)
print('a estimado:', beta[0][0])
print('b estimado:', beta[1][0])
```

```python

plt.figure(figsize=(10,5))
plt.plot(x, y, 'o', label='dados originais')
plt.plot(x,x.dot(beta), label='regressão linear')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Regressão Linear com numpy')
plt.grid()
plt.show()
```
