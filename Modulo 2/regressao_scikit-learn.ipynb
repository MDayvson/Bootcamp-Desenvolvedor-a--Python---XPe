```python
# DADOS

x = [-1, -0.77777778, -0.5555556, -0.3333333, -0.11111111,
     0.111111, 0.33333333, 0.5555556, 0.77777778, 1]

y = [-1.139556201, -0.57177999, -0.21697092, 0.542599, 0.49406657,
     1.1472239, 1.642228553, 2.17499824, 2.64773614, 2.95684202]
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# plot dos dados

plt.figure(figsize=(10,5))
plt.plot(x, y, 'o', label='dados originais')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

```python
# transformando em numpy array

x = np.array(x).reshape(-1, 1)
y = np.array(y).reshape(-1, 1)
```

```python
# Modelo

from sklearn.linear_model import LinearRegression
```

```python
# treinando o modelo: y = a*x + b

reg = LinearRegression()
reg.fit(x, y)
```

```python
# coeficientes estimados

print('a estimado:', reg.coef_.ravel()[0])
print('b estimado:', reg.intercept_[0])
```

```python
# predição do modelo
y_pred = reg.predict(x)
```

```python
# score do modelo

score = reg.score(x, y)
print('Score:', score)
```

```python
plt.figure(figsize=(10,5))
plt.plot(x, y, 'o', label='dados originais')
plt.plot(x, y_pred, label= f'regressão linear (R2:{score:.3f})')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Regressão Linear no scikit-learn')
plt.grid()
plt.show()
```
